```python
df_diet_age_gender = df_food.groupby(["diet_group", "sex", "age_group"])[
    ["mean_ghgs", "mean_land", "mean_watscar", "mean_eut",
     "mean_ghgs_ch4", "mean_ghgs_n2o", "mean_bio", "mean_watuse", "mean_acid"]
].mean().reset_index()

df_diet_age_gender.to_csv("diet_age_gender.csv", index=False)
```

```python
df_diet = df_food.groupby(["diet_group"])[
    ["mean_ghgs", "mean_land", "mean_watscar", "mean_eut",
     "mean_ghgs_ch4", "mean_ghgs_n2o", "mean_bio", "mean_watuse", "mean_acid"]
].mean().reset_index()

df_diet.to_csv("diet.csv", index=False)
```
